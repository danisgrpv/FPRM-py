```python
import numpy as np
import matplotlib.pyplot as plt

# 
L = 20          #  
N = 1000        # 
lambda_ = 0.5   #  
d = 1.0         #   
r_cell = 1.0    #  

#  
x = np.linspace(-L/2, L/2, N)
y = np.linspace(-L/2, L/2, N)
X, Y = np.meshgrid(x, y)

#  
source1_pos = (-d/2, 0)
source2_pos = (d/2, 0)

#     
r1 = np.sqrt((X - source1_pos[0])**2 + (Y - source1_pos[1])**2)
r2 = np.sqrt((X - source2_pos[0])**2 + (Y - source2_pos[1])**2)

#   
delta = 2 * np.pi * (r1 - r2) / lambda_

#    ()
cell_effect = np.where(X**2 + Y**2 <= r_cell**2, 0.5, 1.0)  #    

#  
intensity = (1 + np.cos(delta)) / 2 * cell_effect  #     

# 
plt.figure(figsize=(8, 8))
plt.imshow(intensity, extent=(-L/2, L/2, -L/2, L/2), cmap='gray')
plt.colorbar(label='')
plt.title('  ')
plt.xlabel('X')
plt.ylabel('Y')
```

```python
res = np.fft.fft2(intensity)
res = np.fft.fftshift(res)
a = np.abs(res)
b = np.angle(res)
```

```python
c = np.atan(np.imag(res) / np.real(res))
```

```python
# 
plt.figure(figsize=(8, 8))
plt.imshow(c, cmap='gray')
plt.colorbar(label='')
plt.title('  ')
plt.xlabel('X')
plt.ylabel('Y')
```

```python
import numpy as np
import matplotlib.pyplot as plt

#   
L = 10             #   (   -10  10)
N = 512            # 
x = np.linspace(-L, L, N)
y = np.linspace(-L, L, N)
X, Y = np.meshgrid(x, y)

#     
lambda_ = 0.5     #   (,  )
d = 2.0           #   

#   
source1_pos = (-d/2, 0)
source2_pos = (d/2, 0)

#      
r1 = np.sqrt((X - source1_pos[0])**2 + (Y - source1_pos[1])**2)
r2 = np.sqrt((X - source2_pos[0])**2 + (Y - source2_pos[1])**2)

#   
delta = 2 * np.pi * (r1 - r2) / lambda_

#   
intensity = (1 + np.cos(delta)) / 2  #   

#  
spectrum = np.fft.fftshift(np.fft.fft2(intensity))  #    
spectrum_magnitude = np.abs(spectrum)                #  
spectrum_phase = np.angle(spectrum)                  #  

# 
plt.figure(figsize=(12, 8))

#  
plt.subplot(1, 3, 1)
plt.imshow(intensity, extent=(-L, L, -L, L), cmap='nipy_spectral', origin='lower')
plt.colorbar(label='')
plt.title(' ')
plt.xlabel('X')
plt.ylabel('Y')

#  
plt.subplot(1, 3, 2)
plt.imshow(np.log1p(spectrum_magnitude), extent=(-N/2, N/2, -N/2, N/2), cmap='gray', origin='lower')
plt.colorbar(label=' ')
plt.title('  ()')
plt.xlabel(' X')
plt.ylabel(' Y')

#  
plt.subplot(1, 3, 3)
plt.imshow(spectrum_phase, extent=(-N/2, N/2, -N/2, N/2), cmap='gray', vmin=-np.pi, vmax=np.pi, origin='lower')
plt.colorbar(label='')
plt.title('  ()')
plt.xlabel(' X')
plt.ylabel(' Y')

plt.tight_layout()
```
