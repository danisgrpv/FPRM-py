```python
import numpy as np
import matplotlib.pyplot as plt

# Параметры
L = 20          # размер области
N = 1000        # разрешение
lambda_ = 0.5   # длина волны
d = 1.0         # расстояние между источниками
r_cell = 1.0    # радиус лейкоцита

# Создание сетки
x = np.linspace(-L/2, L/2, N)
y = np.linspace(-L/2, L/2, N)
X, Y = np.meshgrid(x, y)

# Параметры источников
source1_pos = (-d/2, 0)
source2_pos = (d/2, 0)

# Расчет разностей путей от источников
r1 = np.sqrt((X - source1_pos[0])**2 + (Y - source1_pos[1])**2)
r2 = np.sqrt((X - source2_pos[0])**2 + (Y - source2_pos[1])**2)

# Расчет разности фаз
delta = 2 * np.pi * (r1 - r2) / lambda_

# Эффект от лейкоцита (циркуль)
cell_effect = np.where(X**2 + Y**2 <= r_cell**2, 0.5, 1.0)  # уменьшение интенсивности под лейкоцитом

# Интерференционная картина
intensity = (1 + np.cos(delta)) / 2 * cell_effect  # нормализация и эффект от лейкоцита

# Визуализация
plt.figure(figsize=(8, 8))
plt.imshow(intensity, extent=(-L/2, L/2, -L/2, L/2), cmap='gray')
plt.colorbar(label='Интенсивность')
plt.title('Интерферограмма от лейкоцита')
plt.xlabel('X')
plt.ylabel('Y')
```

```python
res = np.fft.fft2(intensity)
res = np.fft.fftshift(res)
a = np.abs(res)
b = np.angle(res)
```

```python
c = np.atan(np.imag(res) / np.real(res))
```

```python
# Визуализация
plt.figure(figsize=(8, 8))
plt.imshow(c, cmap='gray')
plt.colorbar(label='Интенсивность')
plt.title('Интерферограмма от лейкоцита')
plt.xlabel('X')
plt.ylabel('Y')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Параметры интерференционного изображения
L = 10             # размер области (в нашем случае -10 до 10)
N = 512            # разрешение
x = np.linspace(-L, L, N)
y = np.linspace(-L, L, N)
X, Y = np.meshgrid(x, y)

# Длина волны и другие параметры
lambda_ = 0.5     # длина волны (например, в микрометрах)
d = 2.0           # расстояние между источниками

# Положение двух источников
source1_pos = (-d/2, 0)
source2_pos = (d/2, 0)

# Расчет разности путей для двух источников
r1 = np.sqrt((X - source1_pos[0])**2 + (Y - source1_pos[1])**2)
r2 = np.sqrt((X - source2_pos[0])**2 + (Y - source2_pos[1])**2)

# Расчет разности фаз
delta = 2 * np.pi * (r1 - r2) / lambda_

# Генерация интерференционного изображения
intensity = (1 + np.cos(delta)) / 2  # нормализированное значение интенсивности

# Преобразование Фурье
spectrum = np.fft.fftshift(np.fft.fft2(intensity))  # Быстрое двумерное преобразование Фурье
spectrum_magnitude = np.abs(spectrum)                # Модуль спектра
spectrum_phase = np.angle(spectrum)                  # Фаза спектра

# Визуализация
plt.figure(figsize=(12, 8))

# Интерференционная картина
plt.subplot(1, 3, 1)
plt.imshow(intensity, extent=(-L, L, -L, L), cmap='nipy_spectral', origin='lower')
plt.colorbar(label='Интенсивность')
plt.title('Интерференционная картина')
plt.xlabel('X')
plt.ylabel('Y')

# Модуль спектра
plt.subplot(1, 3, 2)
plt.imshow(np.log1p(spectrum_magnitude), extent=(-N/2, N/2, -N/2, N/2), cmap='gray', origin='lower')
plt.colorbar(label='Модуль спектра')
plt.title('Спектр Фурье (модуль)')
plt.xlabel('Частота X')
plt.ylabel('Частота Y')

# Фазовый спектр
plt.subplot(1, 3, 3)
plt.imshow(spectrum_phase, extent=(-N/2, N/2, -N/2, N/2), cmap='gray', vmin=-np.pi, vmax=np.pi, origin='lower')
plt.colorbar(label='Фаза')
plt.title('Спектр Фурье (фаза)')
plt.xlabel('Частота X')
plt.ylabel('Частота Y')

plt.tight_layout()
```
