```python
import numpy as np
import scipy as sc
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Параметры капли
R = 1  # радиус капли

# Создание сетки для координат
phi = np.linspace(0, np.pi, 100)
theta = np.linspace(0, 2 * np.pi, 100)
phi, theta = np.meshgrid(phi, theta)

# Расчет координат x, y, z
x = R * np.sin(phi) * np.cos(theta)
y = R * np.sin(phi) * np.sin(theta)
z = R * np.cos(phi)

# Визуализация капли
fig, ax = plt.subplots(figsize=(7, 7), subplot_kw={'projection': '3d'})
ax.plot_surface(x, y, z, color='b', alpha=0.6)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Костян мега супер реальный кабан!!!')
```

```python
# Параметры
L = 10  # размер области
N = 1024  # разрешение
lambda_ = 0.5  # длина волны
d = 1.0  # расстояние между источниками

# Создание сетки
x = np.linspace(-L/2, L/2, N)
y = np.linspace(-L/2, L/2, N)
X, Y = np.meshgrid(x, y)

# Расчет разности путей от источников
r1 = np.sqrt((X + d/2)**2 + Y**2)
r2 = np.sqrt((X - d/2)**2 + Y**2)

# Расчет разности фаз
delta = 2 * np.pi * (r1 - r2) / lambda_

# Интерференционная картина
intensity = (1 + np.cos(delta)) / 2  # нормализация

# Визуализация
plt.figure(figsize=(8, 8))
plt.imshow(intensity, extent=(-L/2, L/2, -L/2, L/2), cmap='gray')
plt.colorbar(label='Интенсивность')
plt.title('Интерференционная картина')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
